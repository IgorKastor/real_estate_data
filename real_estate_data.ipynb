## Аннотация
**Данные.** Архив объявлений о продаже квартир в Санкт-Петербурге и соседних населенных пунктах за несколько лет взяты из сервиса [Яндекс.Недвижимость](https://realty.ya.ru/). 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных, например, расстояние до центра, аэропорта, ближайшего парка и водоема.

Список параметров:

`total_images` - количество фото в объявлении;     
`last_price` - цена объекта;           
`total_area` - общая площадь;      
`first_day_exposition` - дата размещения объявления;
`rooms` - количество комнат;                   
`ceiling_height` - высота потолков;        
`floors_total` - общее количество этажей в доме;      
`living_area` - жилая площадь;           
`floor` - этаж;                  
`is_apartment` - является ли жилье апартаментами;       
`studio` - является ли жилье студией;              
`open_plan` - квартира открытой планировки;             
`kitchen_area` - площадь кухни;       
`balcony` - площадь балкона;               
`locality_name` - населенный пункт;          
`airports_nearest` - расстояние до ближайшего аэропорта в метрах;  
`cityCenters_nearest` - расстояние до центра города в метрах;   
`parks_around3000` - количество парков в радиусе 3 км;    
`parks_nearest` - расстояние до ближайшего парка;         
`ponds_around3000` - количество водоемов в радиусе 3 км;      
`ponds_nearest` - расстояние до ближайшего водоема в метрах;          
`days_exposition` - число дней размещения объявления до момента снятия

## Предобработка данных

### Нахождение явных  пропусков

### Заполнение пропущенных значений там, где это возможно

```python
#ceiling_height

# нулевых значений нет, но почти в 39% записей высота потолка не указана, найдем медиану
med_ceiling_height = df['ceiling_height'].median()
df['ceiling_height'].describe()
```

Пропуски заполнять не будем, чтобы не исказить статистику. В силу особенностей питерской архитектуры, в центре может быть знаачительное число квартир с потолками выше 3 м. 

```python
# floors_total

# записей, в которых не указано общее число этажей, очень мало - менее 0,4%, поэтому просто удалим их
df.dropna(subset=['floors_total'], inplace=True)
```

```python
# living_area     

# жилая площадь не указана в 8% записей, при этом жилая площадь и общая площадь, где они есть, везде ненулевые
# получим ряд значений доли жилой площади от площади всей квартиры
liv_ar_rates = df['living_area'] / df['total_area'] 
liv_ar_rates.head()
```

```python
# найдем медианную долю жилой площади от площади всей квартиры 
liv_ar_med_percent = liv_ar_rates.median() 
liv_ar_med_percent
```

```python
# заменим пропуски жилой площади cоответствующим значением, 
# полученным из общей площади и медианной доли жилой площади
df['living_area'] = df['living_area'].fillna(value=liv_ar_med_percent * df['total_area'])
```

```python
# is_apartment

# выше видно, что пропущено 88% значений и еще 11% значение нулевые (False)
# проверим, есть ли вообще в датасете апартаменты
df[df['is_apartment'] == True].shape[0]
```

```python
# апартаментов всего 50 из более, чем 23 тыс квартир, - это 0,2%, но данные специфические,
# т.к. касаются относительно редкого типа квартир, они еще могут пригодиться
# поэтому столбец удалять не будем, а пропущенные значения заменим на False
df['is_apartment'] = df['is_apartment'].fillna(value=False)
```

`kitchen_area`          

Пропущено почти 10% значений, при этом нулевая площадь кухни нигде не указана. Поэтому логично предположить, что значение пропускали, если кухни нет вообще или она объединена, например, с гостиной. Обычно такое бывает, если квартира маленькая. 

Проверим это, найдя медианы жилой площади и числа комнат у квартир, где данные о площади кухни отсутствуют

```python
df[df['kitchen_area'].isna()]['living_area'].median()
```

```python
df[df['kitchen_area'].isna()]['rooms'].median()
```

```python
# величины действительно небольшие, поэтому будем считать, что у таких квартир кухня имеет нулевую площадь
df['kitchen_area'] = df['kitchen_area'].fillna(value=0)
```

```python
# balcony        

# в 49% записей число балконов не указано и еще в 16% записей оно нулевое
# можно считать, что в случае пропуска балкона действительно нет
# поэтому заменим тут пропуски нулями
df['balcony'] = df['balcony'].fillna(value=0 )
```

```python
# locality_name    

# название населенного пункта пропущено в 0,2% записей, это довольно мало, 
# просто удалим их, т.к. восстанавливать местоположение по косвенным данным 
# для такого малого количества данных считаю нерациональным
df.dropna(subset=['locality_name'], inplace=True)
```

Информация о расстоянии до аэропорта и центра города, количестве прудов и парков пропущена примерно с одинаковой частотой 23%.  Посмотрим, для каких населенных пунктов кроме Петербурга эти данные указаны

```python
columns = ['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'ponds_around3000']
for c in columns:
    good_suburbs = df[df[c].notna() & (df['locality_name'] != 'Санкт-Петербург')]['locality_name'].unique()
    print(good_suburbs)
```

Итак, видно, что данные о расстоянии до аэропорта и центра города, количестве прудов и парков - не пропущены только для одного и того же набора некоторых пригородов Петербурга. Посмотрим сначала, сколько данных у нас про сам Петербург и про остальные пригороды

```python
df[df['locality_name'] == 'Санкт-Петербург'].shape[0] / df.shape[0]
```

```python
df[df['locality_name'].isin(good_suburbs)].shape[0] / df.shape[0]
```

**Между прочим, установили структуру датасета:**
- 66% записей - про Петербург, 
- 11% - про его пригороды c дополнительной информацией о расстояниях до центра, аэропорта, парков и прудов 
- 23% - другие пригороды и населенные пункты Ленобласти, о которых такой информации нет.

Похоже, что 23% пропусков, о которых говорили выше, как раз приходится на последние. Проверим это 

```python
# найдем, какая доля данных в этих 4 столбцах пропущена 
# для Петербурга с ближними пригородами и для остальных (дальних) пригородов

good_suburbs = list(good_suburbs)
# дополним список ближайших пригородов самим Петербургом
good_suburbs.append('Санкт-Петербург')
# список долей пропусков для Петербурга
SPb_null_rates = []
#список долей пропусков для дальних пригородов
other_suburbs_null_rates = []

for c in columns:
    SPb_rate = df[df['locality_name'] == 'Санкт-Петербург'][c].isna().mean()
    SPb_null_rates.append(SPb_rate)
    other_suburbs_rate = df[~df['locality_name'].isin(good_suburbs)][c].isna().mean()
    other_suburbs_null_rates.append(other_suburbs_rate)
    
pd.DataFrame(data={'Санкт-Петербург':SPb_null_rates, 'Пригор. без доп. инф.':other_suburbs_null_rates}, index=columns)
```

Получилось, что почти все пропуски в указанных 4 столбцах приходятся на пригороды (города Ленобласти), для которых информация доп. информация не указана. Для последних пропуски заполнить не получится. Пропуски заполним только для Петербурга и "хороших" пригородов 

```python
# создадим соответствующий столбец 
df['with_add_data'] = df['locality_name'].isin(good_suburbs)
```

```python
# airports_nearest

# нулевых значений почти нет (домов рядом с аэропортом очень мало) 
med_airport_dist = df.loc[df['locality_name'] == 'Санкт-Петербург', 'airports_nearest'].median()
med_airport_dist
```

```python
# по Яндекс.Картам расстояние от Дворцовой площади до aэропорта Пулково - 23 км,
# поэтому полученное значение кажется реалистичным, заполним пропуски медианой
df['airports_nearest'] = df['airports_nearest'].fillna(value=med_airport_dist)
# отмена заполнения для "плохих" пригородов
df.loc[df['with_add_data'] == False, ['airports_nearest']] = np.nan
```

```python
# cityCenters_nearest    

med_center_dist = df.loc[df['locality_name'] == 'Санкт-Петербург', 'cityCenters_nearest'].median()
med_center_dist
```

```python
# учитывая вышеуказанное расстояние от центра до аэропорта, 
# кажется верным тоже заполнить пропуски медианой
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(value=med_center_dist)
# отмена заполнения для "плохих" пригородов
df.loc[df['with_add_data'] == False, ['cityCenters_nearest']] = np.nan
```

```python
# хотелось бы думать, что "хорошие" пригорды - ближние, проверим это
SPb_dist = df[df['with_add_data'] == True][
    ['locality_name', 'cityCenters_nearest']]

SPb_dist.groupby('locality_name').agg('median').sort_values(
    by='cityCenters_nearest', ascending=False
).reset_index()
```

### Устранение дубликатов

```python
# количество и список уникальных названий населенных пунктов
print(df['locality_name'].unique().shape[0])
df['locality_name'].unique()
```

```python
# сначала всюду заменим Ё на Е
df['locality_name'] = df['locality_name'].apply(lambda x: x.replace('ё', 'е'))

# смотрим, насколько уменьшилось число дубликатов
df['locality_name'].unique().shape[0]
```

```python
# извлечение из строки собственно названия населенного пункта и его типа
def loc_type_name_extract(loc_name):
    for c in loc_name:
        if c.isupper():
            if loc_name.index(c) == 0:
                loc_type = 'город'
                loc_pure_name = loc_name
                return (loc_type, loc_pure_name)
            else:
                loc_type = loc_name[:(loc_name.index(c) - 1)]
                loc_pure_name = loc_name[loc_name.index(c):]
                return (loc_type, loc_pure_name)

# пробный запуск функции
print(loc_type_name_extract('поселок городского типа Большая Ижора'))
loc_type_name_extract('Выборг')
```

```python
# создаем столбец с типом населенного пункта
df['locality_type'] = df['locality_name'].apply(lambda x: loc_type_name_extract(x)[0])
# оставляем в исходном названии населенного пункта только его имя
df['locality_name'] = df['locality_name'].apply(lambda x: loc_type_name_extract(x)[1])
```

```python
# обновленное количество и список уникальных собственных названий населенных пунктов
loc_names = df['locality_name'].unique()
print(loc_names.shape[0])
loc_names
```

Видно, что записей про населенные пункты типа "коттеджный поселок" или "садовое товарищество" всего лишь 6. Будем их тоже относить к поселкам 

```python
# унифицируем типы населенных пунктов
def cut_loc_type(row):
    if 'поселок' in row or 'товарищество' in row:
        return 'поселок'
    elif 'деревня' in row or 'село' in row:
        return 'деревня/село'
    else:
        return 'город'
```

```python
df['locality_type'] = df['locality_type'].apply(cut_loc_type)
```

```python
# обновленный список уникальных типов населенных пунктов
df['locality_type'].unique()
```

```python
# проверим, что каждый населенный пункт относится только к одному типу
dup = (df[['locality_name', 'locality_type']]
          .drop_duplicates()
          .duplicated(keep=False, subset='locality_name')
       )
dup_index = dup[dup == True].index
dup_list = df.loc[dup_index, ['locality_name', 'locality_type']].sort_values(by='locality_name')
```

Значение действительно увеличилось с 0,65 до 0,70. Однако неизвестно, линейная ли зависимость на самом деле, возможно,  корреляция зависит от дополнительных переменных (например, от числа комнат, места расположения квартиры, года постройки дома и т.д.).

Все это требует дополнительного подробного анализа, на который уже нет времени (и сил). 

В качестве примера посмотрим еще  коэффициенты корреляции для срезов:
- по числу комнат (ограничимся числом комнат от 1 до 5, т.к. выше видели, что 4-комнатных квартир только 5%, а 5-комнатных - всего 1%);
- по типу этажа

```python
# коэффициенты корреляции цены и площади для срезов по числу комнат 
df_rooms = df.query('last_price_mln < 100 and rooms < 6')[['last_price_mln', 'total_area', 'rooms']]

corr_list_1 = []
for r in [1, 2, 3, 4, 5]:
    corr_list_1.append(
        df_rooms.query('rooms == @r')['last_price_mln']
        .corr(df_rooms.query('rooms == @r')['total_area'])
    )

pd.DataFrame(data={'k_Пирсона':corr_list_1}, index=[1, 2, 3, 4, 5])
```

```python
# коэффициенты корреляции цены и площади для срезов по типу этажа 
df_floor = df.query('last_price_mln < 25')[['last_price_mln', 'total_area', 'floor_type']]

corr_list_2 = []
for f in ['первый', 'последний', 'другой']:
    corr_list_2.append(
        df_floor.query('floor_type == @f')['last_price_mln']
        .corr(df_floor.query('floor_type == @f')['total_area'])
    )

pd.DataFrame(data={'k_Пирсона':corr_list_2}, index=['первый', 'последний', 'другой'])
```

```python
# таблица с данными о цене 
df_round = df.query('1.9 < last_price_mln < 15')[['last_price_mln'] + col_to_corr]
# срок размещения измерим в месяцах
df_round['days_exposition'] = df_round['days_exposition'] / 30 
df_round.rename(columns={'days_exposition':'weeks_exposition'}, inplace=True)
columns = col_to_corr.copy()
columns.pop()
columns.append('weeks_exposition')
# округляем до целых параметры
df_round[columns] = round(df_round[columns])
df_round['expo_week_day'] += 1
df_round.head(10)
```

**Пропуски данных.**  Информация о расстояниях до центра Петербурга, аэропорта и парка была пропущена почти в четверти записей (большинство пропусков было у квартир, расположенных в пригородах). Она бы очень пригодилась для оценки средних цен квартир, располагающихся в различных пригородах. В 14% записей не было информации о количестве дней размещения объявления, удалось установить, что эти пропуски скорее всего соответсвуют объявлениям, которые на момент выгрузки информации из базы еще не были сняты с публикации.

**Аномальные значения.** Чаще всего были связаны с ошибками при вводе данных в форму объявления (перепутали поля ввода, пропустили цифры и т.п.). Большинство из них удалось верно распознать и восстановить, незначительную  часть удалили.

### Разведочный анализ
#### Информация об объектах недвижимости
**Общая площадь, жилая площадь и площадь кухни.** Наиболее часто выставляли на продажу 1-комнатные квартиры площадью 25-50 кв.м. Для 2-комнатных квартир выделяются 2 подкластера, которые скорее всего соответсвуют типам хрущевской и более современной планировок. Квартир с жилой площадью 20-25 кв.м. очень мало - вероятно, сказывается скачок из-за особенностей планировки при переходе от 1-комнатных к 2-комнатным квартирам. Еще более выражена кластеризация по площади кухни - много отдельных пиков очевидно соответствуют стандартным площадям, заложенным в планировке (например, 6 и 10 кв.м.)

**Цена квартир и 1 кв.м.** Медиана цен 1-комнатных квартир - 3,6 млн руб, 2-комнатных - 4,8 млн, 3-комнатных - 6,2 млн. Медиана цены 1 кв. м у 1-комнатных квартир - 100 тыс руб, у 2-, 3- и 4-комнатных - на 10 тыс дешевле. 

**Число комнат, высота потолков, этажность.** Две трети квартир - однокомнатные или двухкомнатные. Высота потолков в более, чем 90% случаев - 2,5-2,8 м. Более 60% квартир расположены не выше 5-этажа. На 1 этаже располагается 12% квартир, на последнем - 14%. Более 60% домов - не выше 12 этажей, в том числе 25% - пятиэтажки и 16% - девятиэтажки.

**Расстояние до центра города, аэропорта и парков.** Среднее расстояние до центра Петербурга у квартир, расположенных в самом городе, - 12 км, у квартир в  пригородах, о которых есть дополнительные данные, - 30 км. Среднее расстояние до ближайшего аэропорта у квартир в Петербурге и таких пригородах отличается несущественно (28 и 33 км соответственно). Среднее расстояние до ближайшего парка у квартир в Петербурге и в этих пригородах также отличается не существенно и по косвенным оценкам после заполнения пропусков составляет 2-3 км (без учета пропусков - лишь 500 м). По другим пригородам о всех этих параметрах информации не было.

#### Информация об объявлении
**День и месяц публикации объявления.** В будние дни объявления публиковали примерно с одинаковой чвстотой, а в выходные - примерно в 2 раза реже. Реже всего объявления размещали в январе, декабре и мае, чаще всего - с февраля по апрель (в мае - уже в 2 раза реже, чем в предыдущие 3 месяца). С июня по ноябрь частота публикаций равномерно нарастала до февральских значений, затем в декабре-январе падала в 1,5-2 раза до майских значений. (Наименьшее количество публикаций в январе и мае может быть связано с большим количеством выходных и праздничных дней, приходящихся на эти месяцы).

**Скорость продажи квартир.** В среднем продажа квартиры занимает около 3 месяцев (медиана 95 дней). Долгими стоит считать продажи, занявшие более 8 месяцев (3-й квартиль 231 день). Среди последних половина продаж все же произошла в течение следующих 6 месяцев. Аномально долгие продажи - в сроке более 1 г 5 мес. Обращают на себя внимание также всплески в середине 2017 г. и затем в феврале 2018 г. (Они могут быть связаны как с некими актуальными политическими или экономическими событиями, так и с расширением клиентской базы сервиса).

#### Поиск закономерностей
**Факторы, влияющие на цену квартиры.** Наиболее значимо цена квартиры зависела от ее площади (заметная корреляция, k_Пирсона=0,65). При увеличении количества комнат эта корреляция усиливается (k_Пирсона=0,66 у 1-комнатных квартир и k_Пирсона=0,75 у 5-комнатных). Также корреляция сильнее у квартир, расположенных не на первом или последнем этаже (k_Пирсона=0,80).

*Средняя цена квартиры* имеет следующие особенности:
- цена почти равномерно возрастает при увеличении роста общей площади, жилой площади, площади кухни и числа комнат;
- есть тенденция к росту цены при увеличении номера этажа до 10-12, затем цена медленно снижается;
- при удалении от центра цена более-менее равномерно снижается;
- при наличии более 5 фото квартиры в объявлении цена равномерно увеличивается с ростом числа фото;
- цена не зависит от дня недели, в который размещали объявление;
- самые низкие цены были у объявлений, размещенных в мае-июне, самые высокие - в апреле;
- цены снижались в 2014-2017 гг., затем снова начали расти;
- цена прямо не зависела от срока размещения объявления, но при увеличении срока - колебания цены усиливались.

**Цена 1 кв. м. в различных населенных пунктах.** Самая высокая цена 1 кв. м. (90-100 тыс р.) была в самом Петербурге, Пушкине, Кудрово и Парголово. Самая низкая (60-70 тыс р.) - в Выборге и Гатчине. Цена кв. м в пригороде/Ленобласти в среднем на треть дешевле, чем в Петербурге. В городах и поселках, о которых в объявлении содержалась дополнительная информация, средняя цена 1 кв. м была почти в 2 раза выше, чем в тех, о которых дополнительных сведений не сообщалось (а именно, 74-82 против 36-40 тыс р.). Возможно, большинство городов и поселков с более дорогими кв. м все-таки находится ближе к Петербургу, однако полных данных об удаленности от центра у нас нет и проверить это не получится!

**Cредняя цена каждого километра от центра Петербурга.** Цена кв.м отрицательно коррелирует с расстоянием до центра Петербурга. Для квартир, расположенных в самом городе, корреляция эта слабая или умеренная (k_Пирсона = -0,3). Для квартир, расположенных в пригородах, зависимость почти отсутствует (k_Пирсона = -0,1), однако последнее утверждение скорее всего не является статистически значимым в силу относительно малого количества квартир, про которые имелись данные о расстоянии до центра. 

Квартиры, расположенные не дальше 8 км от центра в отношении цены кв.м образуют отдельный кластер со своими закономерностями, что обусловлено характером исторической застройки в центре города. При параллельном предложении цена 1 кв. м в пригородах/Ленобласти, в среднем лишь на 11% ниже, чем в самом Петербурге. 

В пригородных поселках Парголово, Шушары и Стрельна цена кв. м более чем в 2 раза ниже, чем у квартир в Петербурге, расположенных на том же расстоянии до центра. В отдаленных пригородах есть квартиры, где цена кв. м такая же, как в центре, но их можно отнести к статистическим выбросам
